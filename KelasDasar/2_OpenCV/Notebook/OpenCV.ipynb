## **📁 Sesi 1: Membedah Gambar Menjadi Matriks Angka**

### **1.1 Anatomi Gambar Digital: Piksel dan Resolusi**

Setiap gambar digital terbentuk dari jutaan kotak-kotak kecil yang sangat rapat yang disebut **Piksel** (*Picture Element*). Setiap piksel menyimpan informasi kecerahan atau warna dalam bentuk angka.

* **Resolusi Gambar:** Total jumlah piksel yang membentuk lebar dan tinggi gambar. Sebagai contoh, gambar beresolusi $1920 \times 1080$ berarti memiliki lebar 1.920 piksel dan tinggi 1.080 piksel, dengan total kumulatif mencapai $2.073.600$ titik piksel. Komputer melihat gambar ini sebagai sebuah grid raksasa berisi jutaan angka-angka intensitas cahaya.

### **1.2 Sistem Koordinat Gambar Komputer**

Berbeda dengan sistem koordinat Kartesius yang diajarkan pada matematika sekolah (di mana titik $(0,0)$ berada tepat di tengah-tengah), **Sistem Koordinat Digital Komputer menetapkan titik pusat $(0,0)$ mutlak pada ujung Kiri Atas gambar**.

* Sumbu **X** akan bergerak membesar ke arah **Kanan**.
* Sumbu **Y** akan bergerak membesar ke arah **Bawah**.

Pemahaman koordinat ini sangat krusial, karena posisi indeks pemotongan ataupun penggambaran objek visual akan selalu dihitung mulai dari pojok kiri atas sebagai titik nol.

### **1.3 Ruang Warna (Color Space): Grayscale vs BGR/RGB**

Warna direpresentasikan oleh komputer melalui dimensi lapis matriks (atau disebut *Channel*):

1. **Grayscale (1 Channel):** Gambar hitam-putih. Hanya memiliki 1 lapis matriks, di mana setiap nilai piksel berkisar antara `0` (Hitam pekat) hingga `255` (Putih murni). Angka di antaranya melambangkan variasi warna abu-abu.
2. **BGR / RGB (3 Channels):** Gambar berwarna. Memiliki 3 lapis matriks yang ditumpuk secara paralel. Setiap lapis mewakili intensitas warna primer: **Red (Merah)**, **Green (Hijau)**, dan **Blue (Biru)**.

* **Catatan Penting OpenCV:** Secara default, library OpenCV membaca urutan lapis warna terbalik, yaitu **BGR** (Blue, Green, Red), bukan RGB. Oleh karena itu, saat menampilkan gambar OpenCV menggunakan library Matplotlib, kita wajib membalik urutan lapisnya ke RGB agar warna gambar tidak terlihat aneh/biru.

## **📊 Sesi 2: Operasi Geometri & Transformasi Warna**

### **2.2 Mengapa Algoritma AI Lebih Suka Gambar Grayscale?**

Di dalam industri *Computer Vision*, sebelum sebuah gambar dimasukkan ke dalam algoritma pendeteksi bentuk atau wajah, gambar berwarna biasanya diubah terlebih dahulu menjadi Grayscale. Mengapa?

* **Efisiensi Komputasi:** Gambar berwarna memiliki 3 matriks ($Width \times Height \times 3$). Gambar Grayscale hanya memiliki 1 matriks ($Width \times Height \times 1$). Dengan mengubah ke Grayscale, kita memotong beban kerja prosesor hingga **66.6%** tanpa kehilangan informasi bentuk fisik objek dari gambar tersebut.

### **2.2 Efek Resize dan Distorsi Pikselasi**

Mengubah ukuran gambar (*Resize*) adalah proses interpolasi matematis untuk menambah atau mengurangi jumlah piksel.

* Memperkecil gambar terlalu ekstrem akan membuang informasi detail secara permanen.
* Memperbesar gambar di luar resolusi aslinya akan memicu efek **Pikselasi/Pecah**, karena komputer dipaksa menebak dan mengisi kekosongan nilai piksel baru berdasarkan piksel terdekat di sekitarnya.

### **2.3 Cropping Menggunakan Teknik Slicing NumPy**

Karena OpenCV membaca gambar murni sebagai matriks array NumPy (`numpy.ndarray`), kita tidak membutuhkan fungsi khusus untuk memotong gambar (*Cropping*). Kita cukup menerapkan teknik **Array Slicing** yang sudah dipelajari pada Hari 1.

* Format Slicing Matriks Gambar 2D: `gambar[y_awal:y_akhir, x_awal:x_akhir]`
* Ingat: Tentukan rentang sumbu vertikal (tinggi/Y) terlebih dahulu, baru kemudian diikuti rentang sumbu horizontal (lebar/X).

## **📏 Sesi 3: Filter & Deteksi Tepi (Edge Detection)**

### **3.1 Konsep Filter: Mengapa Perlu Blur Sebelum Deteksi?**

Di dunia nyata, gambar sering kali memiliki gangguan visual seperti tekstur bintik-bintik kasar, debu, atau variasi pencahayaan tajam yang disebut **Noise**. Jika gambar penuh *noise* ini langsung diproses, komputer akan menganggap bintik-bintik pengganggu tersebut sebagai objek penting.

* **Analogi Blur:** Seperti saat Anda menyipitkan mata untuk melihat garis besar sebuah papan baliho di kejauhan agar mata Anda tidak terdistraksi oleh detail-detail kecil atau debu yang berterbangan di udara.
* **Gaussian Blur:** Metode pemulasan gambar yang menghitung rata-rata nilai piksel tetangga dengan bobot distribusi lonceng Gaussian. Hasilnya adalah efek buram yang halus, yang efektif membuang bintik *noise* frekuensi tinggi namun tetap mempertahankan struktur utama objek.

### **3.2 Mekanisme Deteksi Tepi (Canny Edge Detection)**

Bagaimana mata komputer mengenali batas akhir suatu benda? Jawabannya adalah lewat **Perubahan Nilai Piksel yang Drastis (Gradien Ekstrem)**.

* Misal ada perbatasan antara objek baju hitam (nilai piksel mendekati `10`) dan latar belakang tembok putih (nilai piksel mendekati `240`). Loncatan nilai dari 10 ke 240 yang terjadi secara instan dalam 1 atau 2 piksel tetangga menandakan adanya **Tepi (Edge)**.
* Algoritma **Canny Edge Detection** mengukur gradien intensitas tersebut secara otomatis, menekan piksel non-maksimum, dan menghasilkan sketsa garis hitam-putih murni yang dengan jelas memperlihatkan siluet kontur dari objek.
