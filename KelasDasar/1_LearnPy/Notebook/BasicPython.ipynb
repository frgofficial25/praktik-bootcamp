## **📁 Sesi 1: Fundamental, Tipe Data, & Struktur Data**

### **1.1 Bagaimana Python Bekerja?**

Sebelum menulis kode, kita harus tahu apa yang terjadi di balik layar. Python adalah bahasa *interpreted*. Saat Anda menjalankan file `.py`, Python interpreter akan menerjemahkan baris kodemu menjadi **Bytecode** (disimpan di folder `__pycache__` sebagai file `.pyc`). Komputer tidak paham bahasa manusia, dan *Bytecode* inilah yang dibaca oleh *Python Virtual Machine* (PVM) agar instruksi kita bisa dieksekusi oleh prosesor.

### **1.2 Variabel, Tipe Data, dan Mekanisme Casting**

**Variabel** adalah nama penanda sebuah lokasi di memori komputer untuk menyimpan nilai data. Di Python, variabel bersifat dinamis (tidak perlu mendeklarasikan tipe datanya secara manual).

Tipe data dasar yang wajib dipahami:

* **`String` (Teks):** Karakter yang diapit kutip. Python mendukung *multiline string* (`"""`) dan *Literal String* (`f"..."`) untuk mempermudah format teks.
* **`Integer` (Angka Bulat):** Bilangan tanpa desimal (Contoh: `10`, `-3`).
* **`Float` (Angka Desimal):** Bilangan pecahan dengan titik desimal (Contoh: `3.14`).
* **`Boolean` (Logika):** Hanya bernilai `True` atau `False`.

**Mekanisme Casting:** Kadang data yang kita terima dari pengguna lewat fungsi `input()` selalu bertipe *String*. **Casting** adalah proses memaksa suatu tipe data berubah menjadi tipe data lain (misal: mengubah `"100"` menjadi `100` menggunakan fungsi `int()`).

### **1.3 Operator Komparasi Kompleks & Bitwise**

Komputer mengambil keputusan berdasarkan perbandingan. Selain operator matematika biasa (`+`, `-`, `*`, `/`), Python memiliki operator komparasi khusus:

* **Operator `is` dan `is not`:** Berbeda dengan `==` yang hanya memeriksa *nilai* objek, operator `is` digunakan untuk memeriksa apakah dua variabel menunjuk ke **alamat memori (object identity) yang sama**.

**Operator Bitwise:** Operator yang melakukan komputasi langsung di tingkat biner (bit per bit). Digunakan untuk optimasi performa tingkat tinggi pada algoritma data:

* `&` (AND), `|` (OR), `^` (XOR), `~` (NOT), serta Shifting (`<<`, `>>`).

### **1.4 Struktur Data Lanjutan: List, Tuple, Set, dan Dictionary**

Ketika mengelola ribuan data, variabel tunggal tidak akan cukup. Kita membutuhkan koleksi data:

1. **`List` (Keranjang Belanja Dinamis):** Koleksi data berurutan menggunakan tanda `[]`. Elemennya bisa dimanipulasi, ditambah (`.append()`), atau disisipkan (`.insert()`). Python juga mendukung *Nested List* (List di dalam List) yang menjadi cikal bakal visualisasi tabel data.
2. **`Tuple` & `Set`:** `Tuple` `()` bersifat *immutable* (isinya dikunci dan tidak bisa diubah setelah dibuat), sangat aman untuk data konstan. `Set` `{}` adalah kumpulan data unik tanpa duplikasi dan tidak berurutan.
3. **`Dictionary` (Buku Kontak Key-Value):** Koleksi data menggunakan kurung kurawal `{}` yang berpasangan. Data dicari berdasarkan **Key** (Kata Kunci), bukan nomor indeks. Sangat cocok untuk memetakan database profil atau biodata terstruktur.


### **💻 Praktik 1: Variabel, Tipe Data, & Mekanisme Casting**

Sebelum membuat program yang kompleks, kita harus terbiasa dulu dengan
"bahan baku" paling dasar dalam pemrograman: variabel dan tipe data.

Di praktik ini kita akan:
- Membuat variabel dengan berbagai tipe data (String, Integer, Float, Boolean)
- Memeriksa tipe data sebuah variabel menggunakan `type()`
- Mempraktikkan *casting* — mengubah satu tipe data menjadi tipe data lain

```python
# Variabel bertipe String (teks)
nama = "Andi"

# Variabel bertipe Integer (angka bulat)
umur = 20

# Variabel bertipe Float (angka desimal)
tinggi_badan = 172.5

# Variabel bertipe Boolean (logika True/False)
sudah_lulus = False

# Cek tipe data dari setiap variabel menggunakan type()
print(f"nama          : {nama}          -> tipe data: {type(nama)}")
print(f"umur          : {umur}          -> tipe data: {type(umur)}")
print(f"tinggi_badan  : {tinggi_badan}  -> tipe data: {type(tinggi_badan)}")
print(f"sudah_lulus   : {sudah_lulus}   -> tipe data: {type(sudah_lulus)}")
```

```python
# Multiline string menggunakan tiga tanda kutip
biodata = """
Nama         : Andi
Umur         : 20 tahun
Tinggi Badan : 172.5 cm
"""
print(biodata)

# f-string (Literal String) untuk menyisipkan variabel langsung ke dalam teks
print(f"Halo, nama saya {nama}. Saya berusia {umur} tahun.")
```

```python
# Simulasikan input dari pengguna -- hasil input() SELALU bertipe String
umur_input = "20"
print(f"Sebelum casting -> nilai: {umur_input}, tipe data: {type(umur_input)}")

# Casting String menjadi Integer menggunakan int()
umur_setelah_casting = int(umur_input)
print(f"Setelah casting -> nilai: {umur_setelah_casting}, tipe data: {type(umur_setelah_casting)}")

# Casting Integer menjadi Float
nilai_float = float(umur_setelah_casting)
print(f"Casting ke Float -> nilai: {nilai_float}, tipe data: {type(nilai_float)}")

# Casting angka menjadi String (kebalikannya)
nilai_string = str(umur_setelah_casting)
print(f"Casting ke String -> nilai: {nilai_string}, tipe data: {type(nilai_string)}")
```

```python
# Casting tidak selalu berhasil -- coba casting teks yang bukan angka
teks_invalid = "dua puluh"

try:
    hasil = int(teks_invalid)
    print(f"Berhasil di-casting: {hasil}")
except ValueError as e:
    print(f"❌ Gagal melakukan casting! Error: {e}")
    print("   Penyebab: teks 'dua puluh' tidak bisa diubah langsung menjadi angka.")
```

### **💻 Praktik 2: Operator Komparasi (`==` vs `is`) & Bitwise**

Komputer mengambil keputusan berdasarkan perbandingan. Di praktik ini kita
akan membedah perbedaan antara `==` (membandingkan **nilai**) dan `is`
(membandingkan **identitas objek di memori**), lalu mencoba operator
Bitwise yang bekerja langsung di level biner.

```python
# Dua variabel dengan nilai yang sama
a = 100
b = 100

print(f"a = {a}, b = {b}")
print(f"a == b -> {a == b}   (membandingkan NILAI)")
print(f"a is b -> {a is b}   (membandingkan ALAMAT MEMORI)")
print(f"id(a)  -> {id(a)}")
print(f"id(b)  -> {id(b)}")
```

```python
# Dua List dengan isi yang identik, tapi dibuat secara terpisah
list_a = [1, 2, 3]
list_b = [1, 2, 3]

# List yang menunjuk ke objek yang SAMA di memori
list_c = list_a

print(f"list_a == list_b -> {list_a == list_b}   (nilainya sama)")
print(f"list_a is list_b -> {list_a is list_b}   (objek BEDA di memori)")
print(f"list_a is list_c -> {list_a is list_c}   (objek SAMA, karena list_c = list_a)")

print()
print(f"id(list_a) -> {id(list_a)}")
print(f"id(list_b) -> {id(list_b)}")
print(f"id(list_c) -> {id(list_c)}")
```

```python
x = 12   # dalam biner: 1100
y = 10   # dalam biner: 1010

print(f"x = {x:>3} -> biner: {bin(x)}")
print(f"y = {y:>3} -> biner: {bin(y)}")
print("=" * 35)

print(f"x & y  (AND)   -> {x & y:>3}  | biner: {bin(x & y)}")
print(f"x | y  (OR)    -> {x | y:>3}  | biner: {bin(x | y)}")
print(f"x ^ y  (XOR)   -> {x ^ y:>3}  | biner: {bin(x ^ y)}")
print(f"~x     (NOT)   -> {~x:>3}  | biner: {bin(~x)}")
```

```python
angka = 4   # biner: 100

print(f"angka asli       : {angka} -> biner: {bin(angka)}")
print(f"angka << 1 (Left Shift)  : {angka << 1} -> biner: {bin(angka << 1)}   (nilai x2)")
print(f"angka << 2 (Left Shift)  : {angka << 2} -> biner: {bin(angka << 2)}   (nilai x4)")
print(f"angka >> 1 (Right Shift) : {angka >> 1} -> biner: {bin(angka >> 1)}   (nilai dibagi 2)")
```

### **💻 Praktik 3: List (Keranjang Dinamis) & Tuple (Data Terkunci)**

Di praktik ini kita akan membandingkan dua struktur data berurutan yang
paling sering dipakai: `List` yang fleksibel dan bisa diubah-ubah, serta
`Tuple` yang sifatnya terkunci (*immutable*) dan aman untuk data konstan.
Kita juga akan mencoba *Nested List* — List di dalam List — sebagai
cikal bakal representasi tabel data.

```python
# List awal berisi daftar belanja
belanjaan = ["beras", "telur", "minyak"]
print(f"List awal      : {belanjaan}")

# .append() -> menambah elemen di akhir List
belanjaan.append("gula")
print(f"Setelah append : {belanjaan}")

# .insert(posisi, nilai) -> menyisipkan elemen di posisi tertentu
belanjaan.insert(1, "sabun")
print(f"Setelah insert : {belanjaan}")

# .remove() -> menghapus elemen berdasarkan nilainya
belanjaan.remove("minyak")
print(f"Setelah remove : {belanjaan}")

# Indexing & slicing List
print(f"\nElemen pertama        : {belanjaan[0]}")
print(f"Elemen terakhir        : {belanjaan[-1]}")
print(f"Dua elemen pertama     : {belanjaan[0:2]}")
```

```python
# Nested List -> List yang setiap elemennya juga berupa List
# Setiap baris merepresentasikan satu data: [nama, umur, kota]
data_peserta = [
    ["Andi", 20, "Bandung"],
    ["Sari", 22, "Jakarta"],
    ["Budi", 19, "Surabaya"]
]

print("📋 Tabel Data Peserta:")
print("-" * 35)
for baris in data_peserta:
    nama, umur, kota = baris  # unpacking otomatis
    print(f"Nama: {nama:<6} | Umur: {umur:<3} | Kota: {kota}")
print("-" * 35)

# Mengakses elemen spesifik di dalam Nested List
# data_peserta[1] -> baris kedua, data_peserta[1][0] -> nama di baris kedua
print(f"\nNama peserta baris ke-2 : {data_peserta[1][0]}")
```

```python
# Tuple cocok untuk data yang tidak boleh berubah, misal koordinat tetap
koordinat = (10, 20)
print(f"Tuple koordinat: {koordinat}")
print(f"Tipe data      : {type(koordinat)}")

# Coba ubah isi Tuple -> akan menghasilkan error
try:
    koordinat[0] = 99
except TypeError as e:
    print(f"\n❌ Gagal mengubah Tuple! Error: {e}")
    print("   Penyebab: Tuple bersifat immutable, isinya terkunci permanen.")
```

### **💻 Praktik 4: Set (Data Unik) & Dictionary (Database Key-Value)**

Di praktik terakhir Sesi 1 ini, kita akan mengenal `Set` — struktur data
yang otomatis menghapus duplikasi, dan `Dictionary` — struktur data yang
mencari informasi berdasarkan **Key** (kata kunci), bukan nomor indeks.
Dictionary inilah yang akan kita gunakan untuk menyusun database biodata
peserta secara rapi dan terstruktur.

```python
# List yang sengaja punya data duplikat
kota_asal_peserta = ["Jakarta", "Bandung", "Jakarta", "Surabaya", "Bandung", "Jakarta"]
print(f"List asli (dengan duplikat) : {kota_asal_peserta}")
print(f"Jumlah data                : {len(kota_asal_peserta)}")

# Mengubah List menjadi Set -> duplikat otomatis terhapus
kota_unik = set(kota_asal_peserta)
print(f"\nSet hasil konversi (unik)  : {kota_unik}")
print(f"Jumlah data unik           : {len(kota_unik)}")

print("\n💡 Set sangat berguna untuk menghitung jumlah kategori unik")
print("   dari ribuan data tanpa perlu mengecek satu per satu secara manual.")
```

```python
# Dictionary menggunakan pasangan Key (kata kunci) dan Value (nilai)
biodata_andi = {
    "nama": "Andi",
    "umur": 20,
    "kota": "Bandung",
    "hobi": ["Coding", "Membaca", "Futsal"]  # Value boleh berupa List
}

print("📋 Biodata Peserta:")
print("-" * 30)
# Mengakses nilai menggunakan Key, BUKAN nomor indeks
print(f"Nama : {biodata_andi['nama']}")
print(f"Umur : {biodata_andi['umur']}")
print(f"Kota : {biodata_andi['kota']}")
print(f"Hobi : {', '.join(biodata_andi['hobi'])}")

# Menambah Key baru ke Dictionary yang sudah ada
biodata_andi["status"] = "Aktif"
print(f"\nSetelah menambah key baru: {biodata_andi}")
```

```python
# Database biodata banyak peserta -> Dictionary di dalam Dictionary
database_peserta = {
    "P001": {"nama": "Andi", "umur": 20, "kota": "Bandung"},
    "P002": {"nama": "Sari", "umur": 22, "kota": "Jakarta"},
    "P003": {"nama": "Budi", "umur": 19, "kota": "Surabaya"}
}

print("📋 DATABASE BIODATA PESERTA")
print("=" * 45)

# .items() -> mengambil sepasang Key dan Value sekaligus saat looping
for id_peserta, data in database_peserta.items():
    print(f"\n🆔 ID Peserta : {id_peserta}")
    for key, value in data.items():
        print(f"   {key.capitalize():<6} : {value}")

print("\n" + "=" * 45)
print(f"Total peserta terdaftar : {len(database_peserta)} orang")
```

```python
# Mencari biodata peserta berdasarkan ID -- jauh lebih cepat dibanding
# mencari satu-satu di dalam List biasa
id_dicari = "P002"

# Menggunakan .get() lebih aman dibanding [] karena tidak akan error
# jika Key-nya tidak ditemukan
hasil_cari = database_peserta.get(id_dicari)

if hasil_cari:
    print(f"✅ Data ditemukan untuk ID {id_dicari}:")
    print(f"   Nama : {hasil_cari['nama']}")
    print(f"   Umur : {hasil_cari['umur']}")
    print(f"   Kota : {hasil_cari['kota']}")
else:
    print(f"❌ Data dengan ID {id_dicari} tidak ditemukan.")
```

## **📊 Sesi 2: Control Flow & Fungsi Terapan**

### **2.1 Cabang Keputusan Dinamis (If-Elif-Else)**

**Control Flow** memberikan "otak" bagi program untuk memilih jalur eksekusi kode berdasarkan kondisi logika `True` atau `False`.

* Di Python, blok percabangan ditandai secara mutlak menggunakan **Indentasi** (spasi/tab), bukan tanda kurung kurawal.

### **2.2 Kontrol Perulangan Tingkat Lanjut (Loops Control)**

Perulangan digunakan agar komputer mengeksekusi perintah repetitif secara otomatis:

* **`for` loop:** Iterasi terukur melewati objek koleksi (seperti List atau fungsi `range()`). Kita bisa mengekstrak nilai sekaligus posisinya memanfaatkan fungsi `enumerate()`.
* **`while` loop:** Perulangan bersyarat yang akan berjalan terus selama kondisinya masih terpenuhi.

**Keywords Pengendali Loop:**

* `pass`: Statemen kosong sebagai *placeholder* agar kode tidak error saat belum diisi logika.
* `continue`: Melewati sisa baris kode di dalam loop pada perulangan saat ini dan langsung meloncat ke iterasi berikutnya.
* `break`: Menghentikan paksa seluruh aktivitas perulangan secara instan, meskipun kondisi utama loop masih bernilai benar.

### **2.3 Modularitas Kode: Fungsi Terapan, `*args`, & `kwargs**`

Fungsi (`def`) adalah blok kode mandiri yang dirancang untuk melakukan tugas spesifik. Agar fungsi kita fleksibel menampung jumlah input yang tidak pasti, kita menggunakan fitur parameter tingkat lanjut:

* **`*args` (Arguments):** Menerima input parameter dalam jumlah berapapun dan membungkusnya secara otomatis menjadi sebuah **Tuple**.
* **`kwargs` (Keyword Arguments):** Menerima input parameter acak yang memiliki nama variabel, lalu membungkusnya secara otomatis menjadi sebuah **Dictionary**.

### **💻 Praktik 1: Cabang Keputusan Dinamis (If-Elif-Else)**

`Control Flow` memberi "otak" bagi program untuk memilih jalur eksekusi
berdasarkan kondisi `True` atau `False`. Di praktik ini kita akan membuat
program penentu nilai (grading) — semakin kompleks aturannya, semakin
banyak cabang `elif` yang dibutuhkan.

Ingat: blok kode di Python ditandai dengan **Indentasi**, bukan tanda
kurung kurawal seperti bahasa lain.

```python
# Kondisi paling dasar: apakah nilai sudah lulus standar minimum?
nilai = 75
standar_lulus = 60

if nilai >= standar_lulus:
    print(f"Nilai {nilai} -> LULUS ✅")
else:
    print(f"Nilai {nilai} -> TIDAK LULUS ❌")
```

```python
nilai = 87

if nilai >= 90:
    grade = "A"
elif nilai >= 80:
    grade = "B"
elif nilai >= 70:
    grade = "C"
elif nilai >= 60:
    grade = "D"
else:
    grade = "E"

print(f"Nilai  : {nilai}")
print(f"Grade  : {grade}")
```

```python
# Program penentu kelayakan beasiswa
nilai_akademik = 88
penghasilan_ortu = 2_500_000  # dalam Rupiah
batas_penghasilan = 3_000_000

if nilai_akademik >= 85 and penghasilan_ortu <= batas_penghasilan:
    print("✅ LAYAK menerima beasiswa penuh.")
elif nilai_akademik >= 75 or penghasilan_ortu <= batas_penghasilan:
    print("⚠️  LAYAK menerima beasiswa sebagian.")
else:
    print("❌ TIDAK LAYAK menerima beasiswa.")

print(f"\nDetail -> Nilai akademik: {nilai_akademik}, Penghasilan ortu: Rp{penghasilan_ortu:,}")
```

```python
# Program verifikasi akun sederhana
username_input = "admin"
password_input = "rahasia123"

username_valid = "admin"
password_valid = "rahasia123"

if username_input == username_valid:
    print("✅ Username ditemukan.")
    if password_input == password_valid:
        print("✅ Password benar -> LOGIN BERHASIL")
    else:
        print("❌ Password salah -> LOGIN GAGAL")
else:
    print("❌ Username tidak ditemukan -> LOGIN GAGAL")
```

### **💻 Praktik 2: Kontrol Perulangan & Filter Data Otomatis**

Perulangan (`for` dan `while`) memungkinkan komputer mengeksekusi perintah
repetitif secara otomatis tanpa kita menulis kode yang sama berulang-ulang.
Ditambah dengan keyword pengendali `pass`, `continue`, dan `break`, kita bisa
membuat program penyaring (*filter*) data secara otomatis dari ribuan baris data.

### **💻 Praktik 3: Kalkulator Interaktif dengan While Loop**

Sekarang kita gabungkan semua yang sudah dipelajari — `while` loop,
`if-elif-else`, casting, dan `break` — untuk membuat sebuah program
kalkulator sederhana yang bisa terus digunakan berulang kali sampai
pengguna sendiri yang memutuskan untuk berhenti.

```python
# Definisikan dulu logika operasi matematikanya secara terpisah
def hitung(angka1, operator, angka2):
    """
    Menerima dua angka dan satu operator, mengembalikan hasil perhitungan.
    Jika operator tidak dikenali atau ada error matematika, kembalikan None.
    """
    if operator == "+":
        return angka1 + angka2
    elif operator == "-":
        return angka1 - angka2
    elif operator == "*":
        return angka1 * angka2
    elif operator == "/":
        if angka2 == 0:
            print("❌ Tidak bisa membagi dengan nol!")
            return None
        return angka1 / angka2
    else:
        print(f"❌ Operator '{operator}' tidak dikenali.")
        return None

# Coba fungsi-nya dulu secara manual sebelum dipakai interaktif
print(hitung(10, "+", 5))
print(hitung(10, "/", 0))
print(hitung(10, "%", 5))
```

```python
print("🧮 KALKULATOR SEDERHANA")
print("Operator yang didukung: + - * /")
print("Ketik 'keluar' pada angka pertama untuk mengakhiri program.\n")

while True:
    angka1_input = input("Masukkan angka pertama (atau 'keluar'): ")

    # Cek kondisi keluar dari loop
    if angka1_input.lower() == "keluar":
        print("\n👋 Kalkulator ditutup. Sampai jumpa!")
        break

    operator_input = input("Masukkan operator (+ - * /): ")
    angka2_input   = input("Masukkan angka kedua: ")

    # Casting input String menjadi Float agar bisa dihitung
    try:
        angka1 = float(angka1_input)
        angka2 = float(angka2_input)
    except ValueError:
        print("❌ Input angka tidak valid, coba lagi.\n")
        continue  # lewati sisa kode, ulang dari awal loop

    hasil = hitung(angka1, operator_input, angka2)

    if hasil is not None:
        print(f"✅ Hasil: {angka1} {operator_input} {angka2} = {hasil}\n")
    else:
        print()  # baris kosong saja sebagai pemisah
```

### **💻 Praktik 4: Fungsi Modular dengan `*args` & `**kwargs`**

Terkadang kita tidak tahu pasti berapa banyak input yang akan diterima
sebuah fungsi. Di praktik ini kita akan belajar membuat fungsi yang
fleksibel menggunakan `*args` (menampung input dalam bentuk Tuple) dan
`**kwargs` (menampung input bernama dalam bentuk Dictionary).

```python
# Fungsi biasa -> hanya bisa menerima jumlah parameter yang FIXED
def jumlahkan_tiga(a, b, c):
    return a + b + c

print(f"Jumlah 3 angka: {jumlahkan_tiga(10, 20, 30)}")

# Masalahnya: bagaimana jika kita ingin menjumlahkan 5 angka? 10 angka?
# Kita tidak mungkin membuat fungsi terpisah untuk setiap kemungkinan jumlah input.
```

```python
def jumlahkan_semua(*args):
    """
    *args menerima input dalam jumlah berapapun dan membungkusnya
    secara otomatis menjadi sebuah Tuple.
    """
    print(f"Tipe data args : {type(args)}")
    print(f"Isi args       : {args}")

    total = 0
    for angka in args:
        total += angka
    return total

print(f"Jumlah 3 angka  : {jumlahkan_semua(10, 20, 30)}")
print(f"Jumlah 5 angka  : {jumlahkan_semua(1, 2, 3, 4, 5)}")
print(f"Jumlah 1 angka  : {jumlahkan_semua(99)}")
```

```python
def buat_profil(**kwargs):
    """
    **kwargs menerima input bernama (keyword) dalam jumlah berapapun
    dan membungkusnya secara otomatis menjadi sebuah Dictionary.
    """
    print(f"Tipe data kwargs : {type(kwargs)}")
    print(f"Isi kwargs       : {kwargs}")
    print()

    print("📋 PROFIL PESERTA")
    print("-" * 25)
    for key, value in kwargs.items():
        print(f"{key.capitalize():<10} : {value}")

# Fungsi ini bisa dipanggil dengan jumlah dan nama parameter yang berbeda-beda
buat_profil(nama="Andi", umur=20, kota="Bandung")
print()
buat_profil(nama="Sari", umur=22, kota="Jakarta", hobi="Coding")
```

```python
def buat_laporan_transaksi(nama_pembeli, *barang, **info_tambahan):
    """
    Parameter biasa  -> wajib diisi, urutannya tetap
    *barang          -> daftar barang yang dibeli (jumlah fleksibel)
    **info_tambahan  -> informasi ekstra opsional (metode bayar, diskon, dll)
    """
    print("🧾 LAPORAN TRANSAKSI")
    print("=" * 35)
    print(f"Pembeli       : {nama_pembeli}")

    print(f"Barang dibeli : {', '.join(barang)}")

    if info_tambahan:
        print("Info tambahan :")
        for key, value in info_tambahan.items():
            print(f"   - {key.capitalize()}: {value}")

    print("=" * 35)

# Contoh pemanggilan dengan kombinasi ketiga jenis parameter
buat_laporan_transaksi(
    "Budi",
    "Pulpen", "Buku", "Penghapus",
    metode_bayar="QRIS",
    diskon="10%"
)
```

### **💻 Praktik 4: Operasi Linear Aljabar untuk Gambar (Dot, Cross, Invers)**

Operasi-operasi ini adalah motor penggerak di balik algoritma Computer
Vision — mulai dari transformasi koordinat, proyeksi objek, hingga
membalikkan kembali sebuah transformasi spasial pada gambar.

```python
matriks_a = np.array([
    [1, 2],
    [3, 4]
])

matriks_b = np.array([
    [5, 6],
    [7, 8]
])

print("Matriks A:")
print(matriks_a)
print("\nMatriks B:")
print(matriks_b)

# Perkalian element-wise BIASA (bukan perkalian matriks sebenarnya)
print("\nA * B (element-wise, posisi sejajar dikalikan langsung):")
print(matriks_a * matriks_b)

# Perkalian matriks SEBENARNYA menggunakan .dot()
print("\nA.dot(B) (perkalian matriks / dot product):")
print(matriks_a.dot(matriks_b))
```

```python
# Bayangkan ini adalah satu titik koordinat (x, y) pada sebuah gambar
titik_koordinat = np.array([4, 3])

# Matriks transformasi untuk memperbesar gambar 2x (scaling)
matriks_scaling = np.array([
    [2, 0],
    [0, 2]
])

titik_baru = matriks_scaling.dot(titik_koordinat)

print(f"Koordinat asli         : {titik_koordinat}")
print(f"Matriks transformasi (scaling 2x):")
print(matriks_scaling)
print(f"\nKoordinat setelah transformasi : {titik_baru}")
print("\n💡 Inilah dasar bagaimana AI memperbesar, memutar, atau memindahkan")
print("   posisi objek pada sebuah gambar secara matematis.")
```

```python
# Cross product hanya berlaku pada vektor 3D, sering dipakai untuk
# menghitung arah normal (tegak lurus) pada permukaan objek 3D
vektor_a = np.array([1, 0, 0])
vektor_b = np.array([0, 1, 0])

hasil_cross = np.cross(vektor_a, vektor_b)

print(f"Vektor A          : {vektor_a}")
print(f"Vektor B          : {vektor_b}")
print(f"Cross Product A×B : {hasil_cross}")
print("\n💡 Cross product menghasilkan vektor baru yang tegak lurus terhadap")
print("   kedua vektor asalnya — dipakai untuk menghitung arah normal permukaan")
print("   pada rendering objek 3D.")
```

```python
# Matriks transformasi scaling 2x dari contoh sebelumnya
matriks_scaling = np.array([
    [2, 0],
    [0, 2]
])

# Hitung invers-nya menggunakan np.linalg.inv()
matriks_invers = np.linalg.inv(matriks_scaling)

print("Matriks transformasi asli (scaling 2x):")
print(matriks_scaling)
print("\nMatriks invers (untuk membalikkan transformasi):")
print(matriks_invers)

# Buktikan: titik yang sudah ditransformasi, dikalikan invers, akan KEMBALI ke asal
titik_asli           = np.array([4, 3])
titik_setelah_scaling = matriks_scaling.dot(titik_asli)
titik_kembali_normal  = matriks_invers.dot(titik_setelah_scaling)

print(f"\nTitik koordinat asli         : {titik_asli}")
print(f"Setelah di-scaling 2x        : {titik_setelah_scaling}")
print(f"Setelah dikembalikan (invers): {titik_kembali_normal}")
print("\n💡 Invers matriks dipakai untuk membalikkan transformasi spasial,")
print("   misalnya mengembalikan gambar yang sudah diputar atau dicerminkan ke posisi semula.")
```

### **4.1 Pilar Utama OOP: Class, Object, dan Variabel Instance**

```python
class RekeningBank:
    def __init__(self, pemilik, saldo_awal):
        self.pemilik = pemilik
        self.__saldo = saldo_awal

    @property
    def saldo(self):
        """Getter -- cara AMAN untuk MEMBACA nilai variabel privat."""
        return self.__saldo

    @saldo.setter
    def saldo(self, nilai_baru):
        """Setter -- cara AMAN untuk MENGUBAH nilai variabel privat, dengan validasi."""
        if nilai_baru < 0:
            print("❌ Saldo tidak boleh bernilai negatif! Perubahan ditolak.")
            return
        self.__saldo = nilai_baru
        print(f"✅ Saldo berhasil diperbarui menjadi Rp{self.__saldo:,}")

rekening = RekeningBank("Andi", 500_000)

# Mengakses nilai lewat Getter (terlihat seperti atribut biasa, tapi sebenarnya method)
print(f"Saldo saat ini : Rp{rekening.saldo:,}")

# Mengubah nilai lewat Setter -- otomatis tervalidasi
rekening.saldo = 750_000
print(f"Saldo terbaru  : Rp{rekening.saldo:,}")

# Coba ubah ke nilai negatif -- akan ditolak oleh validasi Setter
rekening.saldo = -100_000
print(f"Saldo tetap    : Rp{rekening.saldo:,}")
```

### **💻 Praktik 3: Inheritance, `super()`, & Method Resolution Order (MRO)**

`Inheritance` (pewarisan) memungkinkan sebuah Class baru (*Child Class*)
mewarisi seluruh atribut dan method dari Class induknya (*Parent Class*),
sehingga kita tidak perlu menulis ulang kode yang sama. Kita juga akan
melihat bagaimana Python menentukan urutan eksekusi saat sebuah Class
mewarisi dari lebih dari satu induk (*Multiple Inheritance*).

```python
class Hewan:
    """Parent Class -- berisi atribut dan method umum yang dimiliki semua hewan."""

    def __init__(self, nama, suara):
        self.nama = nama
        self.suara = suara

    def bersuara(self):
        print(f"{self.nama} bersuara: {self.suara}!")

    def info(self):
        print(f"Ini adalah {self.nama}, sejenis hewan.")


class Kucing(Hewan):
    """Child Class -- otomatis mewarisi __init__, bersuara(), dan info() dari Hewan."""
    pass


class Anjing(Hewan):
    pass

# Kedua Child Class ini langsung bisa pakai method dari Parent Class
kucing = Kucing("Kitty", "Meong")
anjing = Anjing("Brownie", "Guk")

kucing.bersuara()
kucing.info()

anjing.bersuara()
anjing.info()
```

```python
class Kucing(Hewan):
    def __init__(self, nama, warna_bulu):
        # super() memanggil __init__ milik Parent Class (Hewan)
        # agar kita tidak perlu menulis ulang logika nama & suara dari awal
        super().__init__(nama, suara="Meong")
        self.warna_bulu = warna_bulu  # atribut TAMBAHAN khusus milik Kucing

    def info(self):
        """Override -- menimpa method info() milik Parent dengan versi baru."""
        super().info()  # tetap panggil versi Parent dulu sebagai dasar
        print(f"Warna bulunya {self.warna_bulu}.")

kucing = Kucing("Kitty", "Oren")
kucing.bersuara()  # method ini TIDAK di-override, tetap pakai milik Parent
kucing.info()      # method ini SUDAH di-override, perilakunya berubah
```

```python
class Karyawan:
    def bekerja(self):
        print("Sedang menjalankan tugas kantor.")


class Mahasiswa:
    def belajar(self):
        print("Sedang mengerjakan tugas kuliah.")


class KaryawanSambilKuliah(Karyawan, Mahasiswa):
    """Multiple Inheritance -- mewarisi dari DUA Parent Class sekaligus."""
    pass

orang = KaryawanSambilKuliah()
orang.bekerja()   # method dari Karyawan
orang.belajar()   # method dari Mahasiswa
```

```python
class A:
    def info(self):
        print("Method info() milik Class A")

class B(A):
    def info(self):
        print("Method info() milik Class B")

class C(A):
    def info(self):
        print("Method info() milik Class C")

class D(B, C):
    """D mewarisi dari B dan C, yang keduanya juga mewarisi dari A -- inilah Diamond Problem."""
    pass

objek_d = D()
objek_d.info()  # method mana yang dijalankan? cek urutan MRO di bawah

# __mro__ menunjukkan urutan pencarian method dari kiri ke kanan
print("\nUrutan MRO Class D:")
for index, kelas in enumerate(D.__mro__, start=1):
    print(f"{index}. {kelas.__name__}")

print("\n💡 Python mencari method secara berurutan sesuai MRO: D -> B -> C -> A -> object")
print("   Itulah mengapa info() yang dijalankan adalah milik Class B, bukan C.")
```

### **💻 Praktik 4: Magic Methods & Abstract Class**

`Magic Methods` adalah method bawaan Python yang diawali dan diakhiri
*double underscore* (seperti `__str__` atau `__add__`). Dengan
memodifikasinya, Object buatan kita bisa merespons perintah bawaan Python
seperti `print()` atau operator matematika `+`. Kita juga akan mengenal
`Abstract Class` — cetakan induk yang tidak bisa dicetak langsung menjadi
Object, dan memaksa semua Class anak untuk mengimplementasikan fungsi tertentu.

```python
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

produk1 = Produk("Pulpen", 3000)

# Tanpa __str__, print() pada Object hanya menampilkan info memori yang tidak berguna
print(produk1)
print("\n⚠️  Outputnya tidak informatif sama sekali untuk pengguna.")
```

```python
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

    def __str__(self):
        """Dipanggil otomatis saat Object di-print() -- representasi untuk manusia."""
        return f"Produk: {self.nama} (Rp{self.harga:,})"

    def __repr__(self):
        """Representasi teknis, biasanya untuk debugging di console."""
        return f"Produk(nama='{self.nama}', harga={self.harga})"

produk1 = Produk("Pulpen", 3000)

print(produk1)          # otomatis memanggil __str__
print(repr(produk1))    # otomatis memanggil __repr__
```

```python
class Produk:
    def __init__(self, nama, harga):
        self.nama = nama
        self.harga = harga

    def __str__(self):
        return f"{self.nama} (Rp{self.harga:,})"

    def __add__(self, lainnya):
        """Memodifikasi perilaku operator '+' antar dua Object Produk."""
        total_harga = self.harga + lainnya.harga
        return f"Total harga {self.nama} + {lainnya.nama} = Rp{total_harga:,}"

    def __eq__(self, lainnya):
        """Memodifikasi perilaku operator '==' -- bandingkan berdasarkan harga."""
        return self.harga == lainnya.harga

    def __lt__(self, lainnya):
        """Memodifikasi perilaku operator '<' -- untuk keperluan sorting."""
        return self.harga < lainnya.harga

pulpen  = Produk("Pulpen", 3000)
pensil  = Produk("Pensil", 3000)
penghapus = Produk("Penghapus", 2000)

print(pulpen + pensil)               # memanggil __add__
print(f"pulpen == pensil -> {pulpen == pensil}")     # memanggil __eq__
print(f"penghapus < pulpen -> {penghapus < pulpen}") # memanggil __lt__

# Karena __lt__ sudah didefinisikan, sorted() otomatis bisa dipakai!
daftar_produk = [pulpen, pensil, penghapus]
daftar_terurut = sorted(daftar_produk)
print("\nDaftar produk terurut dari harga termurah:")
for produk in daftar_terurut:
    print(f"- {produk}")
```

```python
from abc import ABC, abstractmethod

class BentukGeometri(ABC):
    """
    Abstract Class -- class suci ini TIDAK BISA dicetak langsung menjadi Object.
    Hanya berfungsi sebagai kontrak cetak biru bagi seluruh Class anak.
    """

    @abstractmethod
    def hitung_luas(self):
        """Setiap Child Class WAJIB mengimplementasikan method ini sendiri."""
        pass

# Coba cetak Object langsung dari Abstract Class -> akan error
try:
    bentuk = BentukGeometri()
except TypeError as e:
    print(f"❌ Gagal membuat Object! Error: {e}")
    print("   Penyebab: BentukGeometri adalah Abstract Class, tidak bisa diinstansiasi langsung.")
```

```python
class Persegi(BentukGeometri):
    def __init__(self, sisi):
        self.sisi = sisi

    def hitung_luas(self):
        """WAJIB diimplementasikan, karena ini abstractmethod dari Parent."""
        return self.sisi ** 2


class Lingkaran(BentukGeometri):
    def __init__(self, radius):
        self.radius = radius

    def hitung_luas(self):
        return 3.14 * (self.radius ** 2)

persegi = Persegi(5)
lingkaran = Lingkaran(3)

print(f"Luas Persegi (sisi=5)    : {persegi.hitung_luas()}")
print(f"Luas Lingkaran (radius=3): {lingkaran.hitung_luas():.2f}")

print("\n💡 Abstract Class memaksa setiap Child Class untuk punya method hitung_luas(),")
print("   tapi cara perhitungannya bisa berbeda-beda sesuai bentuk masing-masing.")
```

### **💻 Praktik 5 (Mini Project): Kasir Toko & Inventaris Mandiri**

Di Mini Project ini kita akan menggabungkan **seluruh konsep OOP** yang
sudah dipelajari — Class, Enkapsulasi, Inheritance, dan Magic Methods —
menjadi sebuah aplikasi kasir toko yang interaktif dan fungsional.

Arsitektur aplikasi:
- `Produk` → Class dasar yang merepresentasikan satu item barang
- `Inventaris` → Class yang mengelola keseluruhan stok barang
- `KasirToko` → Child Class dari Inventaris, menangani transaksi belanja

```python
class Produk:
    def __init__(self, kode, nama, harga, stok):
        self.kode  = kode
        self.nama  = nama
        self.__harga = harga   # Private -- harga tidak boleh diubah sembarangan
        self.__stok  = stok    # Private -- stok hanya boleh berkurang saat transaksi

    @property
    def harga(self):
        return self.__harga

    @property
    def stok(self):
        return self.__stok

    def kurangi_stok(self, jumlah):
        if jumlah > self.__stok:
            print(f"❌ Stok {self.nama} tidak cukup! Stok tersedia: {self.__stok}")
            return False
        self.__stok -= jumlah
        return True

    def tambah_stok(self, jumlah):
        if jumlah <= 0:
            print("❌ Jumlah penambahan stok harus lebih dari 0.")
            return
        self.__stok += jumlah
        print(f"✅ Stok {self.nama} berhasil ditambah {jumlah}. Stok sekarang: {self.__stok}")

    def __str__(self):
        return f"[{self.kode}] {self.nama:<15} Rp{self.__harga:>8,}  | Stok: {self.__stok}"

    def __repr__(self):
        return f"Produk(kode='{self.kode}', nama='{self.nama}', harga={self.__harga}, stok={self.__stok})"
```

```python
class Inventaris:
    def __init__(self, nama_toko):
        self.nama_toko = nama_toko
        self.__daftar_produk = {}   # Private -- {kode: Object Produk}

    def tambah_produk(self, produk):
        if produk.kode in self.__daftar_produk:
            print(f"⚠️  Kode {produk.kode} sudah ada di inventaris.")
            return
        self.__daftar_produk[produk.kode] = produk
        print(f"✅ Produk '{produk.nama}' berhasil ditambahkan ke inventaris.")

    def cari_produk(self, kode):
        return self.__daftar_produk.get(kode, None)

    def tampilkan_inventaris(self):
        print(f"\n{'=' * 55}")
        print(f"{'📦 INVENTARIS TOKO: ' + self.nama_toko:^55}")
        print(f"{'=' * 55}")
        print(f"{'Kode':<6} {'Nama':<15} {'Harga':>12}  {'Stok':>6}")
        print(f"{'─' * 55}")
        for produk in self.__daftar_produk.values():
            print(produk)
        print(f"{'=' * 55}")
        print(f"Total jenis produk: {len(self.__daftar_produk)}")
```

```python
class KasirToko(Inventaris):
    def __init__(self, nama_toko):
        super().__init__(nama_toko)
        self.__keranjang   = {}    # {kode: jumlah_beli}
        self.__total_bayar = 0
        self.__riwayat_transaksi = []

    def tambah_ke_keranjang(self, kode, jumlah):
        produk = self.cari_produk(kode)
        if produk is None:
            print(f"❌ Produk dengan kode '{kode}' tidak ditemukan.")
            return

        berhasil = produk.kurangi_stok(jumlah)
        if not berhasil:
            return

        if kode in self.__keranjang:
            self.__keranjang[kode] += jumlah
        else:
            self.__keranjang[kode] = jumlah

        subtotal = produk.harga * jumlah
        self.__total_bayar += subtotal
        print(f"🛒 {produk.nama} x{jumlah} ditambahkan ke keranjang. Subtotal: Rp{subtotal:,}")

    def tampilkan_keranjang(self):
        if not self.__keranjang:
            print("🛒 Keranjang masih kosong.")
            return

        print(f"\n{'=' * 55}")
        print(f"{'🛒 KERANJANG BELANJA':^55}")
        print(f"{'=' * 55}")
        print(f"{'Nama':<18} {'Harga':>10} {'Qty':>5} {'Subtotal':>12}")
        print(f"{'─' * 55}")

        for kode, jumlah in self.__keranjang.items():
            produk   = self.cari_produk(kode)
            subtotal = produk.harga * jumlah
            print(f"{produk.nama:<18} {produk.harga:>10,} {jumlah:>5} {subtotal:>12,}")

        print(f"{'─' * 55}")
        print(f"{'TOTAL':>40} {self.__total_bayar:>12,}")
        print(f"{'=' * 55}")

    def proses_pembayaran(self, uang_bayar):
        if not self.__keranjang:
            print("❌ Keranjang kosong! Tidak ada yang bisa dibayar.")
            return

        if uang_bayar < self.__total_bayar:
            print(f"❌ Uang tidak cukup! Kurang: Rp{self.__total_bayar - uang_bayar:,}")
            return

        kembalian = uang_bayar - self.__total_bayar

        print(f"\n{'=' * 55}")
        print(f"{'✅ PEMBAYARAN BERHASIL':^55}")
        print(f"{'=' * 55}")
        print(f"  Total belanja : Rp{self.__total_bayar:>10,}")
        print(f"  Uang bayar    : Rp{uang_bayar:>10,}")
        print(f"  Kembalian     : Rp{kembalian:>10,}")
        print(f"{'=' * 55}")
        print("  Terima kasih sudah berbelanja! 🙏")
        print(f"{'=' * 55}\n")

        # Simpan ke riwayat transaksi
        self.__riwayat_transaksi.append({
            "keranjang" : dict(self.__keranjang),
            "total"     : self.__total_bayar,
            "bayar"     : uang_bayar,
            "kembalian" : kembalian
        })

        # Reset keranjang untuk transaksi berikutnya
        self.__keranjang   = {}
        self.__total_bayar = 0

    def tampilkan_riwayat(self):
        if not self.__riwayat_transaksi:
            print("📋 Belum ada riwayat transaksi.")
            return

        print(f"\n{'=' * 55}")
        print(f"{'📋 RIWAYAT TRANSAKSI':^55}")
        print(f"{'=' * 55}")
        for i, transaksi in enumerate(self.__riwayat_transaksi, start=1):
            print(f"\nTransaksi #{i}")
            print(f"  Total bayar : Rp{transaksi['total']:,}")
            print(f"  Kembalian   : Rp{transaksi['kembalian']:,}")
        print(f"\n{'=' * 55}")
        print(f"Total transaksi selesai: {len(self.__riwayat_transaksi)}")
        print(f"{'=' * 55}")
```

```python
# Buat Object KasirToko
toko = KasirToko("Toko Nalara")

# Buat beberapa produk
produk_list = [
    Produk("P001", "Pulpen",     3_000,  50),
    Produk("P002", "Buku Tulis", 15_000, 30),
    Produk("P003", "Penghapus",  2_000,  40),
    Produk("P004", "Pensil",     2_500,  60),
    Produk("P005", "Penggaris",  5_000,  20),
]

# Masukkan semua produk ke inventaris toko
for produk in produk_list:
    toko.tambah_produk(produk)

# Tampilkan inventaris awal
toko.tampilkan_inventaris()
```

```python
print("🛍️  TRANSAKSI 1 — Pelanggan: Andi\n")

toko.tambah_ke_keranjang("P001", 3)   # 3 Pulpen
toko.tambah_ke_keranjang("P002", 2)   # 2 Buku Tulis
toko.tambah_ke_keranjang("P003", 1)   # 1 Penghapus

toko.tampilkan_keranjang()
toko.proses_pembayaran(uang_bayar=50_000)
```

```python
print("🛍️  TRANSAKSI 2 — Pelanggan: Sari\n")

toko.tambah_ke_keranjang("P005", 5)    # 5 Penggaris
toko.tambah_ke_keranjang("P002", 50)   # stok tidak cukup, harusnya ditolak

toko.tampilkan_keranjang()
toko.proses_pembayaran(uang_bayar=30_000)
```

```python
# Restok salah satu produk
print("📦 RESTOK BARANG")
produk_pulpen = toko.cari_produk("P001")
produk_pulpen.tambah_stok(100)

# Inventaris setelah transaksi & restok
toko.tampilkan_inventaris()

# Riwayat seluruh transaksi yang sudah selesai
toko.tampilkan_riwayat()
```
